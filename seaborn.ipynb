```python
#Multiple plot
```

```python
df=sns.load_dataset("exercise")

df.head()
```

```python
mp=sns.FacetGrid(df, col="kind", hue="kind")

mp.map(sns.scatterplot, "time", "pulse")

mp.add_legend()
```

```python
mp=sns.FacetGrid(df, col="kind", hue="time")

mp.map(sns.barplot, "time", "pulse")

mp.add_legend()
```
